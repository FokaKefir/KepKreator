# Import Letters dataset

```python
from emnist import extract_training_samples
x_letter_train, y_letter_train = extract_training_samples('letters')
x_letter_train = x_letter_train / 255.0
```

```python
x_letter_train.shape, y_letter_train.shape
```

```python
y_letter_train.min(), y_letter_train.max()
```

```python
def decode_letter(num: int):
  return chr(96 + num)
def encode_letter(letter: str):
  return ord(letter) - 96
```

```python
random_indices = np.random.choice(x_letter_train.shape[0], 9, replace=False)

plt.figure(figsize=(10, 10))
for i, index in enumerate(random_indices, 1):
    plt.subplot(3, 3, i)
    plt.imshow(x_letter_train[index], cmap='gray')
    plt.title(f"Label: {decode_letter(y_letter_train[index])}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# Import ByClass dataset

```python
from emnist import extract_training_samples
x_train, y_train = extract_training_samples('byclass')
x_train = x_train / 255.0
```

```python
x_train.shape, y_train.shape
```

```python
y_train.min(), y_train.max()
```

```python
def decode_byclass(num: int):
  if num < 10:
    return str(num)
  elif 10 <= num <= 35:
    return chr(num - 10 + 65)
  else:
    return chr(num - 36 + 97)
```

```python
inds = []
for label in range(y_train.max() + 1):
  index = np.where(y_train == label)[0][0]
  inds.append(index)
```

```python
plt.figure(figsize=(10, 15))
for i, index in enumerate(inds, 1):
    plt.subplot(10, 7, i)
    plt.imshow(x_train[index], cmap='gray')
    plt.title(f"Label: {y_train[index]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
random_indices = np.random.choice(x_train.shape[0], 9, replace=False)

plt.figure(figsize=(10, 10))
for i, index in enumerate(random_indices, 1):
    plt.subplot(3, 3, i)
    plt.imshow(x_train[index], cmap='gray')
    plt.title(f"Label: {decode_byclass(y_train[index])}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
